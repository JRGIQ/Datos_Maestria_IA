````python
def factorial(n):
  """Calcula el factorial de un número entero no negativo."""
  if n < 0:
    return "El factorial no está definido para números negativos"
  elif n == 0:
    return 1
  else:
    result = 1
    for i in range(1, n + 1):
      result *= i
    return result

# Ejemplo de uso:
numero = 5
print(f"El factorial de {numero} es {factorial(numero)}")

numero = 0
print(f"El factorial de {numero} es {factorial(numero)}")

numero = -3
print(f"El factorial de {numero} es {factorial(numero)}")
````

````python
#función que calcule el factorial de un numero entero positivo
# 5!=5*4*3*2*1=120
# 1!=1 0!=1
import time

#Factorial iterativo
def facto_i(n):
  if n<0:
    return 'ERROR'
  elif n==0:
    return 1
  var=1
  # Esta es la iteración
  for i in range(1,n+1):
    var=var*i
  return var

inicio=time.time()
var=facto_i(978)
fin=time.time()
print(var)
print('El tiempo es de:', fin-inicio)
````

````python
#Factorial recursivo
def facto_r(n):
  if n < 0:
    return 'ERROR'
  elif n==0 or n==1:
    return 1
  else:
    return n*facto_r(n-1)

inicio=time.time()
var=facto_r(978)
fin=time.time()
print(var)
print('El tiempo es de:', fin-inicio)
print(facto_r(978))
````

# Prueba del algoritmo de expansión de NODOS


````python
from collections import deque
import heapq
from dataclasses import dataclass, field
from typing import Any, Iterable, List, Optional, Tuple, Dict, Set, Callable

# -----------------------------
# Estructuras y utilidades base
# -----------------------------

@dataclass(order=True)
class PrioritizedItem:
    priority: int
    item: Any=field(compare=False)

@dataclass
class Node:
    state: Any
    parent: Optional['Node']
    action: Optional[str]
    path_cost: int
    depth: int

    def solution(self) -> List[str]:
        """Reconstruye la secuencia de acciones desde la raíz hasta este nodo."""
        actions = []
        n = self
        while n and n.action is not None:
            actions.append(n.action)
            n = n.parent
        return list(reversed(actions))

class Fringe:
    """Interfaz mínima: push, pop, empty."""
    def push(self, node: Node, priority: Optional[int]=None): ...
    def pop(self) -> Node: ...
    def empty(self) -> bool: ...

class StackFringe(Fringe):
    def __init__(self):
        self._data: List[Node] = []
    def push(self, node: Node, priority: Optional[int]=None):
        self._data.append(node)
    def pop(self) -> Node:
        return self._data.pop()
    def empty(self) -> bool:
        return not self._data

class QueueFringe(Fringe):
    def __init__(self):
        self._data: deque[Node] = deque()
    def push(self, node: Node, priority: Optional[int]=None):
        self._data.append(node)
    def pop(self) -> Node:
        return self._data.popleft()
    def empty(self) -> bool:
        return not self._data

class PriorityFringe(Fringe):
    """Para UCS; prioridad por path_cost (o lo que se pase)."""
    def __init__(self):
        self._data: List[PrioritizedItem] = []
        self._push_counter = 0  # para desempatar (estable)
    def push(self, node: Node, priority: Optional[int]=None):
        if priority is None:
            priority = node.path_cost
        # añadir contador evita comparar nodos cuando hay empate
        heapq.heappush(self._data, PrioritizedItem((priority, self._push_counter), node))
        self._push_counter += 1
    def pop(self) -> Node:
        return heapq.heappop(self._data).item
    def empty(self) -> bool:
        return not self._data


# -----------------------------------
# Motor genérico de búsqueda en grafos
# -----------------------------------

def graph_search(
    initial_state: Any,
    goal_test: Callable[[Any], bool],
    expand: Callable[[Any], Iterable[Tuple[str, Any, int]]],
    fringe: Fringe,
    use_closed: bool = True
) -> Optional[Node]:
    """
    Implementación fiel del pseudocódigo GRAPH-SEARCH.
    - expand(state) -> iterable de (action, next_state, step_cost)
    - fringe: StackFringe (DFS), QueueFringe (BFS), PriorityFringe (UCS)
    - use_closed: True para evitar reexpansiones de estados visitados
    """
    closed: Set[Any] = set()
    root = Node(initial_state, parent=None, action=None, path_cost=0, depth=0)
    fringe.push(root)

    while True:
        if fringe.empty():
            return None  # failure
        node = fringe.pop()

        if goal_test(node.state):
            return node

        if (not use_closed) or (node.state not in closed):
            if use_closed:
                closed.add(node.state)
            for action, child_state, step_cost in expand(node.state):
                child = Node(
                    state=child_state,
                    parent=node,
                    action=action,
                    path_cost=node.path_cost + step_cost,
                    depth=node.depth + 1
                )
                # Para cola/stack basta push; para prioridad puede pasarse prioridad explícita
                if isinstance(fringe, PriorityFringe):
                    fringe.push(child, priority=child.path_cost)
                else:
                    fringe.push(child)


# ---------------------
# 8-Puzzle (8 piezas)
# ---------------------

# Representamos el tablero como una tupla de 9 enteros, 0 = blanco.
# Índices:
# 0 1 2
# 3 4 5
# 6 7 8

Action = str
State = Tuple[int, ...]  # tupla de 9

MOVES = {
    'Up':    -3,
    'Down':   3,
    'Left':  -1,
    'Right':  1
}

def idx_to_rc(i: int) -> Tuple[int, int]:
    return divmod(i, 3)  # (row, col)

def can_move(blank_idx: int, action: Action) -> bool:
    r, c = idx_to_rc(blank_idx)
    if action == 'Up':    return r > 0
    if action == 'Down':  return r < 2
    if action == 'Left':  return c > 0
    if action == 'Right': return c < 2
    return False

def do_move(state: State, action: Action) -> State:
    blank = state.index(0)
    if not can_move(blank, action):
        raise ValueError("Acción inválida desde este estado")
    offset = MOVES[action]
    target = blank + offset
    s = list(state)
    s[blank], s[target] = s[target], s[blank]
    return tuple(s)

def expandable_8puzzle(order: Tuple[Action, ...]) -> Callable[[State], Iterable[Tuple[Action, State, int]]]:
    """
    Devuelve una función expand(state) que genera sucesores en el orden indicado.
    order: tupla como ('Up','Down','Left','Right') u otra permutación.
    """
    def expand(state: State) -> Iterable[Tuple[Action, State, int]]:
        for a in order:
            blank = state.index(0)
            if can_move(blank, a):
                yield (a, do_move(state, a), 1)  # coste por paso = 1
    return expand

def goal_test_8puzzle(state: State) -> bool:
    return state == (1,2,3,4,5,6,7,8,0)

def parity(state: State) -> int:
    """Paridad de inversiones (para saber si es resoluble)."""
    arr = [x for x in state if x != 0]
    inv = 0
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] > arr[j]:
                inv += 1
    return inv % 2

def is_solvable(state: State) -> bool:
    # En 8-puzzle (3x3), solvable si la paridad de inversiones es par.
    return parity(state) == 0

# -------------------------
# Algoritmos "no informados"
# -------------------------

def breadth_first_search(initial: State, order=('Up','Down','Left','Right')) -> Optional[Node]:
    return graph_search(
        initial_state=initial,
        goal_test=goal_test_8puzzle,
        expand=expandable_8puzzle(order),
        fringe=QueueFringe(),
        use_closed=True
    )

def depth_first_search(initial: State, order=('Up','Down','Left','Right'), depth_limit: Optional[int]=None) -> Optional[Node]:
    """
    DFS con límite opcional de profundidad (DLS).
    Si depth_limit es None, hace DFS sin límite (puede ciclar en estados muy profundos si no hay closed).
    """
    def expand_with_limit(state: State):
        # Usamos un cierre que ve el nodo actual vía una variable mutable externa no disponible:
        # Para respetar GRAPH-SEARCH, haremos el filtro en el bucle principal
        raise NotImplementedError

    # Para conservar GRAPH-SEARCH intacto, envolvemos el expand y descartamos en el push.
    expand = expandable_8puzzle(order)
    closed: Set[State] = set()
    root = Node(initial, None, None, 0, 0)
    fringe = StackFringe()
    fringe.push(root)

    while True:
        if fringe.empty():
            return None
        node = fringe.pop()

        if goal_test_8puzzle(node.state):
            return node

        if node.state not in closed:
            closed.add(node.state)
            if depth_limit is None or node.depth < depth_limit:
                for action, child_state, step_cost in expand(node.state):
                    child = Node(child_state, node, action, node.path_cost + step_cost, node.depth + 1)
                    fringe.push(child)

def iterative_deepening_search(initial: State, order=('Up','Down','Left','Right'), max_depth: int=50) -> Optional[Node]:
    for d in range(max_depth + 1):
        result = depth_first_search(initial, order=order, depth_limit=d)
        if result is not None:
            return result
    return None

def uniform_cost_search(initial: State, order=('Up','Down','Left','Right')) -> Optional[Node]:
    # Con coste uniforme = 1 por paso, UCS genera lo mismo que BFS en número de pasos mínimos.
    return graph_search(
        initial_state=initial,
        goal_test=goal_test_8puzzle,
        expand=expandable_8puzzle(order),
        fringe=PriorityFringe(),
        use_closed=True
    )

# -------------------------
# Ejemplos de ejecución
# -------------------------

def print_solution(node: Node):
    print(f"Profundidad: {node.depth}, costo: {node.path_cost}, pasos: {len(node.solution())}")
    print("Acciones:", node.solution())
    # Traza de estados (opcional)
    path = []
    n = node
    while n:
        path.append(n.state)
        n = n.parent
    for s in reversed(path):
        print_state(s)
        print("")

def print_state(state: State):
    g = [state[i*3:(i+1)*3] for i in range(3)]
    for row in g:
        print(" ".join("{:1}".format(x) if x != 0 else "·" for x in row))

if __name__ == "__main__":
    # Un estado solvible (fácil)
    start: State = (1, 2, 3,
                    4, 5, 6,
                    0, 7, 8)

    # Otro solvible un poco más mezclado
    start2: State = (1, 2, 3,
                     4, 0, 6,
                     7, 5, 8)

    for s in (start, start2):
        print("Estado inicial:")
        print_state(s)
        print("Solvable?", is_solvable(s))
        print("-"*30)

        # Cambia el orden de expansión para observar diferencias:
        order = ('Up','Left','Right','Down')  # prueba otros: ('Left','Right','Up','Down'), etc.

        print("BFS:")
        sol_bfs = breadth_first_search(s, order=order)
        if sol_bfs: print_solution(sol_bfs)
        else: print("Sin solución")

        print("\nDFS (límite 20):")
        sol_dfs = depth_first_search(s, order=order, depth_limit=20)
        if sol_dfs: print_solution(sol_dfs)
        else: print("No encontró solución dentro del límite")

        print("\nIDS:")
        sol_ids = iterative_deepening_search(s, order=order, max_depth=40)
        if sol_ids: print_solution(sol_ids)
        else: print("Sin solución hasta la profundidad indicada")

        print("\nUCS:")
        sol_ucs = uniform_cost_search(s, order=order)
        if sol_ucs: print_solution(sol_ucs)
        else: print("Sin solución")
        print("="*50)
````

# Semana 3
-Comparación de algortimos de búsqueda


````python

# search_benchmark.py
# Framework para comparar algoritmos de búsqueda (informada vs no informada)
# Problemas: 8-puzzle, laberinto, TSP pequeño y mochila 0/1


import time, math, random, heapq, tracemalloc, sys
from dataclasses import dataclass, field
from typing import Any, List, Tuple, Optional, Dict, Set, Iterable

random.seed(42)

@dataclass
class SearchResult:
    problem: str
    algorithm: str
    cost: float
    solution_length: int
    nodes_expanded: int
    time_ms: float
    peak_mem_kb: int
    optimal: Optional[bool]
    extra: Dict[str, Any] = field(default_factory=dict)

class Problem:
    def initial_state(self) -> Any: ...
    def is_goal(self, state: Any) -> bool: ...
    def successors(self, state: Any): ...
    def heuristic(self, state: Any) -> float: return 0.0
    def state_hash(self, state: Any): return state
    def solution_cost(self, path): return sum(c for _,_,c in path)

def measure_run(algorithm_name: str, func, *args, **kwargs):
    tracemalloc.start()
    t0 = time.perf_counter()
    path, nodes, meta = func(*args, **kwargs)
    t1 = time.perf_counter()
    _, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    problem: Problem = kwargs.get("problem") or args[0]
    cost = problem.solution_cost(path) if path is not None else math.inf
    slen = len(path) if path is not None else 0
    return SearchResult(
        problem=meta.get("problem_name", "unknown"),
        algorithm=algorithm_name,
        cost=cost,
        solution_length=slen,
        nodes_expanded=nodes,
        time_ms=(t1-t0)*1000,
        peak_mem_kb=int(peak/1024),
        optimal=meta.get("optimal"),
        extra=meta
    ), path

def bfs(problem: Problem, max_expansions: int = 1_000_000):
    from collections import deque
    start = problem.initial_state()
    if problem.is_goal(start): return [], 0, {"problem_name": type(problem).__name__}
    frontier = deque([(start, [])])
    visited = {problem.state_hash(start)}
    expansions = 0
    while frontier and expansions < max_expansions:
        state, path = frontier.popleft()
        expansions += 1
        for a, ns, c in problem.successors(state):
            h = problem.state_hash(ns)
            if h in visited: continue
            visited.add(h)
            np = path + [(a, ns, c)]
            if problem.is_goal(ns): return np, expansions, {"problem_name": type(problem).__name__, "visited": len(visited)}
            frontier.append((ns, np))
    return None, expansions, {"problem_name": type(problem).__name__, "cutoff": True}

def dfs(problem: Problem, depth_limit: int = 50_000, max_expansions: int = 1_000_000):
    start = problem.initial_state()
    if problem.is_goal(start): return [], 0, {"problem_name": type(problem).__name__}
    stack = [(start, [], 0)]
    visited: Set[Any] = set()
    expansions = 0
    while stack and expansions < max_expansions:
        state, path, depth = stack.pop()
        h = problem.state_hash(state)
        if h in visited: continue
        visited.add(h)
        if depth > depth_limit: continue
        expansions += 1
        for a, ns, c in problem.successors(state):
            np = path + [(a, ns, c)]
            if problem.is_goal(ns): return np, expansions, {"problem_name": type(problem).__name__, "visited": len(visited)}
            stack.append((ns, np, depth+1))
    return None, expansions, {"problem_name": type(problem).__name__, "cutoff": True}

def ucs(problem: Problem, max_expansions: int = 1_000_000):
    start = problem.initial_state()
    pq = [(0.0, 0, start, [])]
    best: Dict[Any, float] = {problem.state_hash(start): 0.0}
    expansions = 0; t=0
    while pq and expansions < max_expansions:
        g, _, s, path = heapq.heappop(pq)
        if problem.is_goal(s): return path, expansions, {"problem_name": type(problem).__name__, "optimal": True}
        expansions += 1
        for a, ns, c in problem.successors(s):
            ng = g+c; h = problem.state_hash(ns)
            if ng < best.get(h, math.inf):
                best[h] = ng; t+=1
                heapq.heappush(pq, (ng, t, ns, path+[(a,ns,c)]))
    return None, expansions, {"problem_name": type(problem).__name__, "cutoff": True}

def greedy_best_first(problem: Problem, max_expansions: int = 1_000_000):
    start = problem.initial_state()
    pq = [(problem.heuristic(start), 0, start, [])]
    seen:set = set(); expansions=0; t=0
    while pq and expansions < max_expansions:
        _,_,s,path = heapq.heappop(pq)
        h = problem.state_hash(s)
        if h in seen: continue
        seen.add(h)
        if problem.is_goal(s): return path, expansions, {"problem_name": type(problem).__name__}
        expansions += 1
        for a, ns, c in problem.successors(s):
            t+=1; heapq.heappush(pq, (problem.heuristic(ns), t, ns, path+[(a,ns,c)]))
    return None, expansions, {"problem_name": type(problem).__name__, "cutoff": True}

def astar(problem: Problem, max_expansions: int = 1_000_000):
    start = problem.initial_state()
    gbest: Dict[Any, float] = {problem.state_hash(start): 0.0}
    pq = [(problem.heuristic(start), 0, start, [])]
    expansions=0; t=0
    while pq and expansions < max_expansions:
        f,_,s,path = heapq.heappop(pq)
        g = gbest.get(problem.state_hash(s), math.inf)
        if problem.is_goal(s): return path, expansions, {"problem_name": type(problem).__name__, "optimal": True}
        expansions += 1
        for a, ns, c in problem.successors(s):
            ng = g+c; h = problem.state_hash(ns)
            if ng < gbest.get(h, math.inf):
                gbest[h]=ng; t+=1
                heapq.heappush(pq, (ng+problem.heuristic(ns), t, ns, path+[(a,ns,c)]))
    return None, expansions, {"problem_name": type(problem).__name__, "cutoff": True}

# 8-Puzzle
class EightPuzzle(Problem):
    goal = (1,2,3,4,5,6,7,8,0)
    def __init__(self, start): self.start=start
    def initial_state(self): return self.start
    def is_goal(self, s): return s==self.goal
    def state_hash(self, s): return s
    @staticmethod
    def _moves(idx):
        r,c = divmod(idx,3)
        for dr,dc,a in [(-1,0,'U'),(1,0,'D'),(0,-1,'L'),(0,1,'R')]:
            nr,nc = r+dr, c+dc
            if 0<=nr<3 and 0<=nc<3: yield a, nr*3+nc
    def successors(self, s):
        i = s.index(0)
        for a,j in self._moves(i):
            lst=list(s); lst[i],lst[j]=lst[j],lst[i]
            yield a, tuple(lst), 1.0
    def heuristic(self, s):
        d=0
        for i,t in enumerate(s):
            if t==0: continue
            gr,gc = divmod(t-1,3); r,c = divmod(i,3)
            d+=abs(gr-r)+abs(gc-c)
        return d

def random_eight_puzzle(steps=12):
    start = EightPuzzle.goal; s=start; prev=None
    import random
    for _ in range(steps):
        i = s.index(0)
        moves = list(EightPuzzle._moves(i))
        if prev is not None: moves=[m for m in moves if m[1]!=prev]
        a,j = random.choice(moves)
        lst=list(s); lst[i],lst[j]=lst[j],lst[i]
        prev=i; s=tuple(lst)
    return s

# Laberinto
class GridMaze(Problem):
    def __init__(self, grid, start, goal):
        self.grid=grid; self.start=start; self.goal=goal
        self.R=len(grid); self.C=len(grid[0])
    def initial_state(self): return self.start
    def is_goal(self, s): return s==self.goal
    def successors(self, s):
        r,c=s
        for dr,dc,a in [(-1,0,'U'),(1,0,'D'),(0,-1,'L'),(0,1,'R')]:
            nr,nc=r+dr,c+dc
            if 0<=nr<self.R and 0<=nc<self.C and self.grid[nr][nc]==0:
                yield a,(nr,nc),1.0
    def heuristic(self, s):
        return abs(s[0]-self.goal[0])+abs(s[1]-self.goal[1])

def random_maze(R=20,C=20,density=0.2):
    import random
    g=[[0]*C for _ in range(R)]
    for r in range(R):
        for c in range(C):
            if (r,c) in [(0,0),(R-1,C-1)]: continue
            if random.random()<density: g[r][c]=1
    g[0][0]=0; g[R-1][C-1]=0
    return g,(0,0),(R-1,C-1)

# Small TSP
class SmallTSP(Problem):
    def __init__(self, coords):
        self.coords=coords; self.nodes=list(coords.keys())
        self.n=len(self.nodes); self.start_node=self.nodes[0]
        self.dist={(i,j): ((coords[i][0]-coords[j][0])**2+(coords[i][1]-coords[j][1])**2)**0.5
                   for i in self.nodes for j in self.nodes}
    def initial_state(self): return (self.start_node, frozenset([self.start_node]), (self.start_node,))
    def is_goal(self, s):
        curr, vis, path = s
        return len(vis)==self.n and curr==self.start_node and len(path)==self.n+1
    def successors(self, s):
        curr, vis, path = s
        if len(vis)==self.n:
            if curr!=self.start_node:
                nxt=self.start_node; cost=self.dist[(curr,nxt)]
                yield f"return({curr}->{nxt})", (nxt, vis, path+(nxt,)), cost
            return
        for nxt in self.nodes:
            if nxt in vis: continue
            cost=self.dist[(curr,nxt)]
            yield f"goto({curr}->{nxt})", (nxt, vis|{nxt}, path+(nxt,)), cost
    def state_hash(self, s):
        curr, vis, path = s
        return (curr, vis)
    def heuristic(self, s):
        curr, vis, path = s
        remaining=[n for n in self.nodes if n not in vis]
        if not remaining: return self.dist[(curr, self.start_node)] if curr!=self.start_node else 0.0
        min_curr=min(self.dist[(curr,r)] for r in remaining)
        min_to_start=min(self.dist[(r,self.start_node)] for r in remaining)
        def mst_cost(nodes):
            if len(nodes)<=1: return 0.0
            used={nodes[0]}; edges=[]
            import heapq
            for j in range(1,len(nodes)):
                heapq.heappush(edges, (self.dist[(nodes[0], nodes[j])], nodes[j]))
            total=0.0
            while len(used)<len(nodes):
                c,v = heapq.heappop(edges)
                if v in used: continue
                total+=c; used.add(v)
                for w in nodes:
                    if w not in used:
                        heapq.heappush(edges, (self.dist[(v,w)], w))
            return total
        return min_curr + mst_cost(remaining) + min_to_start

def random_tsp(n=7, W=100.0, H=100.0):
    import random
    return {i: (random.random()*W, random.random()*H) for i in range(n)}

# Knapsack Branch & Bound
from dataclasses import dataclass

@dataclass
class Item:
    value:int; weight:int

class KnapsackBB(Problem):
    def __init__(self, items: List[Item], capacity:int):
        self.items=items; self.capacity=capacity; self.n=len(items)
    def initial_state(self): return (0,0,0,tuple())
    def is_goal(self, s): idx, w, v, ch = s; return idx==self.n
    def successors(self, s):
        idx, w, v, ch = s
        if idx>=self.n: return
        yield f"skip({idx})", (idx+1, w, v, ch), 0.0
        it=self.items[idx]
        if w+it.weight<=self.capacity:
            yield f"take({idx})", (idx+1, w+it.weight, v+it.value, ch+(idx,)), 0.0
    def heuristic(self, s):
        idx, w, v, ch = s
        cap_left = self.capacity - w
        if cap_left<=0 or idx>=self.n: return 0.0
        rem=self.items[idx:]
        rem_sorted=sorted(rem, key=lambda it: it.value/it.weight, reverse=True)
        bound=0.0; tw=0
        for it in rem_sorted:
            if tw+it.weight<=cap_left:
                tw+=it.weight; bound+=it.value
            else:
                remain=cap_left-tw
                if remain>0: bound += it.value*(remain/it.weight)
                break
        return bound
    def objective_value(self, s):
        idx,w,v,ch = s; return v

def knapsack_dfs_complete(problem: KnapsackBB):
    best=(-1,None); stack=[(problem.initial_state(),[])]; expansions=0
    while stack:
        s,path = stack.pop(); expansions+=1
        if problem.is_goal(s):
            val=problem.objective_value(s)
            if val>best[0]: best=(val,path)
            continue
        for a, ns, c in problem.successors(s):
            stack.append((ns, path+[(a,ns,c)]))
    return best[1], expansions, {"problem_name": type(problem).__name__, "optimal": True, "best_value": best[0], "note": "cost = -max_value"}

def knapsack_best_first_bb(problem: KnapsackBB):
    start=problem.initial_state()
    def prio(s):
        idx,w,v,ch=s; return -(v+problem.heuristic(s))
    pq=[(prio(start),0,start,[])]; best_val=-1; best_path=None; expansions=0; t=0
    while pq:
        _,_,s,path = heapq.heappop(pq); expansions+=1
        if problem.is_goal(s):
            val=problem.objective_value(s)
            if val>best_val: best_val=val; best_path=path
            continue
        idx,w,v,ch=s; optimistic = v+problem.heuristic(s)
        if optimistic<=best_val: continue
        for a, ns, c in problem.successors(s):
            t+=1; heapq.heappush(pq, (prio(ns), t, ns, path+[(a,ns,c)]))
    return best_path, expansions, {"problem_name": type(problem).__name__, "best_value": best_val, "note": "cost = -max_value"}

def run_benchmarks():
    results=[]; details={}
    # Eight puzzle
    start8 = random_eight_puzzle(12)
    p8 = EightPuzzle(start8)
    rA8,_ = measure_run("A*", astar, problem=p8); results.append(rA8); opt8=rA8.cost
    rB8,_ = measure_run("BFS", bfs, problem=p8); results.append(rB8)
    rU8,_ = measure_run("UCS", ucs, problem=p8); results.append(rU8)
    rD8,_ = measure_run("DFS", dfs, problem=p8, depth_limit=60); results.append(rD8)
    rG8,_ = measure_run("Greedy Best-First", greedy_best_first, problem=p8); results.append(rG8)
    for r in results:
        if r.problem=="EightPuzzle":
            r.optimal = (r.algorithm in ("UCS","A*")) or (r.cost==opt8)
    details["eight_puzzle_start"]=start8
    # Maze
    def random_maze(R=20,C=20,density=0.2):
        g=[[0]*C for _ in range(R)]
        import random
        for r in range(R):
            for c in range(C):
                if (r,c) in [(0,0),(R-1,C-1)]: continue
                if random.random()<density: g[r][c]=1
        g[0][0]=0; g[R-1][C-1]=0; return g,(0,0),(R-1,C-1)
    grid,s,g = random_maze(20,20,0.2)
    pM = GridMaze(grid,s,g)
    rBM,_=measure_run("BFS", bfs, problem=pM); results.append(rBM); optM=rBM.cost
    rAM,_=measure_run("A*", astar, problem=pM); results.append(rAM)
    rUM,_=measure_run("UCS", ucs, problem=pM); results.append(rUM)
    rDM,_=measure_run("DFS", dfs, problem=pM, depth_limit=400); results.append(rDM)
    rGM,_=measure_run("Greedy Best-First", greedy_best_first, problem=pM); results.append(rGM)
    for r in results:
        if r.problem=="GridMaze":
            if r.algorithm in ("BFS","UCS","A*") and math.isfinite(r.cost): r.optimal=True
            else: r.optimal=(r.cost==optM)
    details["maze_start_goal"]=(s,g)
    # TSP
    def random_tsp(n=7,W=100.0,H=100.0):
        import random
        return {i:(random.random()*W, random.random()*H) for i in range(n)}
    coords = random_tsp(7)
    pT = SmallTSP(coords)
    rAT,_=measure_run("A*", astar, problem=pT, max_expansions=500_000); results.append(rAT); optT=rAT.cost
    rUT,_=measure_run("UCS", ucs, problem=pT, max_expansions=1_000_000); results.append(rUT)
    rDT,_=measure_run("DFS (Backtracking)", dfs, problem=pT, depth_limit=1000, max_expansions=1_000_000); results.append(rDT)
    rGT,_=measure_run("Greedy Best-First", greedy_best_first, problem=pT, max_expansions=1_000_000); results.append(rGT)
    for r in results:
        if r.problem=="SmallTSP":
            r.optimal = (r.algorithm in ("UCS","A*") and math.isfinite(r.cost)) or (r.cost==optT)
    details["tsp_coords"]=coords
    # Knapsack
    from random import randint
    items=[Item(value=randint(5,40), weight=randint(1,15)) for _ in range(12)]; cap=35
    pK=KnapsackBB(items, cap)
    rDK,_=measure_run("DFS Exhaustive (opt)", knapsack_dfs_complete, problem=pK); results.append(rDK)
    rDK.cost = -rDK.extra.get("best_value", rDK.cost)
    rBK,_=measure_run("Best-First B&B", knapsack_best_first_bb, problem=pK); results.append(rBK)
    rBK.cost = -rBK.extra.get("best_value", rBK.cost)
    best_val=rDK.extra.get("best_value", None)
    for r in results:
        if r.problem=="KnapsackBB":
            r.optimal = (r.extra.get("best_value", None)==best_val)
    details["knapsack_capacity"]=cap; details["knapsack_items"]=[(it.value,it.weight) for it in items]
    return results, details

if __name__=="__main__":
    try:
        import pandas as pd
        res, det = run_benchmarks()
        rows=[{
            "Problema": r.problem, "Algoritmo": r.algorithm,
            "Costo (más bajo es mejor)": round(r.cost,4) if math.isfinite(r.cost) else r.cost,
            "Longitud solución": r.solution_length,
            "Nodos expandidos": r.nodes_expanded,
            "Tiempo (ms)": round(r.time_ms,2),
            "Memoria pico (KB)": r.peak_mem_kb,
            "Óptimo": r.optimal
        } for r in res]
        df=pd.DataFrame(rows)
        order=["EightPuzzle","GridMaze","SmallTSP","KnapsackBB"]
        df["Problema"]=pd.Categorical(df["Problema"], categories=order, ordered=True)
        df=df.sort_values(["Problema","Algoritmo"]).reset_index(drop=True)
        print(df.to_string(index=False))
    except ImportError:
        res, det = run_benchmarks()
        for r in res:
            print(f"{r.problem:12s} | {r.algorithm:20s} | cost={r.cost:.3f} | len={r.solution_length:3d} | "
                  f"expanded={r.nodes_expanded:6d} | time={r.time_ms:.2f}ms | mem={r.peak_mem_kb}KB | optimal={r.optimal}")
````
