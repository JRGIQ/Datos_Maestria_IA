# Procesamiento Básico de Lenguaje Natural: Tokenization, Stemming y Lemmatization

Esta sección del **Diplomado** presenta un análisis de las técnicas básicas de procesamiento de lenguaje natural (PLN), centrado en tres etapas fundamentales: tokenization, stemming y lemmatization. Estas etapas forman parte del preprocesamiento de texto y permiten convertir datos textuales no estructurados en formatos útiles para tareas de análisis y modelado.

## Tokenization

La tokenización es el proceso de dividir un texto en unidades más pequeñas llamadas tokens. Estas unidades pueden ser palabras, oraciones o caracteres, dependiendo del nivel de análisis.

En el documento de estudio se emplea la biblioteca NLTK para realizar tokenización a nivel de oraciones (`sent_tokenize`) y palabras (`word_tokenize`).

**Ejemplo en código:**
```python
from nltk.tokenize import sent_tokenize, word_tokenize
text = 'Hola mundo. Esto es PLN.'
sent_tokenize(text)
word_tokenize(text)
```

Esta etapa es crucial como paso previo a cualquier análisis textual, ya que define la granularidad del procesamiento posterior.

## Stemming

El stemming reduce las palabras a su raíz o forma base (stem), sin garantizar que el resultado sea una palabra válida del idioma.

Se puede considerar el algoritmo de Porter (`PorterStemmer`) proporcionado por NLTK, una herramienta clásica y eficiente para este propósito.

**Ejemplo en código:**
```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmer.stem('running')  # Resultado: 'run'
```

El stemming es útil para reducir la dimensionalidad del vocabulario y agrupar palabras con significados similares, aunque puede producir raíces que no son gramaticalmente correctas.

## Lemmatization

La lematización también busca reducir las palabras a su forma base, llamada lema, pero a diferencia del stemming, garantiza que el resultado sea una palabra válida, considerando su contexto gramatical.

Es una alternativa más precisa al stemming, especialmente en corpus grandes o tareas sensibles al significado.

**Ejemplo en código (usando WordNetLemmatizer):**
```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('running', pos='v')  # Resultado: 'run'
```

La lematización es más costosa computacionalmente, pero preferida en aplicaciones donde la precisión lingüística es prioritaria.

## Comentarios

Las técnicas de tokenización, stemming y lematización son pasos fundamentales en cualquier flujo de trabajo de PLN. La correcta aplicación de estas herramientas depende del tamaño del corpus, la naturaleza de la tarea y el equilibrio entre eficiencia y precisión.

El documento de referencia proporciona ejemplos prácticos de cómo aplicar estas técnicas en Python, lo que permite a los estudiantes desarrollar habilidades sólidas para el análisis de texto automatizado.

#¿Qué es Word Embedding?

Los **Word Embeddings** son una representación numérica de las palabras en un espacio de vectores de varias dimensiones. Esta representación **captura el significado y el contexto** de las palabras, permitiendo que palabras con significados similares estén cerca entre sí en el espacio vectorial.



## ¿Por qué usar Word Embeddings?

Representaciones tradicionales como Bag of Words o TF-IDF solo consideran la **frecuencia** de las palabras, pero no su significado ni similitud.

- En Bag of Words: `"perro"` y `"can"` son diferentes.
- En Word Embeddings: `"perro"` y `"can"` tienen vectores similares porque se usan en contextos parecidos.



## ¿Cómo se construyen?

### 1. **Word2Vec (Google, 2013)**

Modelos principales:
- **CBOW**: Predice una palabra dada su vecindario.
- **Skip-gram**: Predice el contexto a partir de una palabra.

### 2. **GloVe (Stanford)**

Se basa en la **coocurrencia global** de palabras. Usa una matriz de coocurrencias y la factoriza para obtener los vectores.

### 3. **FastText (Facebook)**

Representa palabras como suma de **n-gramas de caracteres**, útil para manejar errores ortográficos y palabras raras.



## Propiedades de los Word Embeddings

- Permiten operaciones aritméticas semánticas:

```plaintext
vec("rey") - vec("hombre") + vec("mujer") ≈ vec("reina")
```

- Capturan relaciones semánticas ("bueno" ≈ "excelente") y sintácticas ("correr" ≈ "corriendo").



## Ejemplo en Python con Gensim (Word2Vec)

```python
from gensim.models import Word2Vec

## Comentarios

- Los Word Embeddings son esenciales para representar palabras de manera semántica.
- Se usan como entrada en modelos de redes neuronales como LSTM, Transformers y BERT.
- Pueden ser preentrenados o entrenados sobre un corpus propio.

````python
import nltk
import gensim
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize, sent_tokenize
import string

# Descargar recursos
nltk.download('punkt')

# Texto de ejemplo en español
texto = """
La inteligencia artificial está revolucionando la forma en que interactuamos con la tecnología.
Muchas empresas utilizan algoritmos para mejorar sus servicios y productos.
El aprendizaje automático es una rama fundamental de la inteligencia artificial.
"""

# Tokenización por oraciones
oraciones = sent_tokenize(texto, language="spanish")

# Preprocesamiento y tokenización por palabras
def limpiar_y_tokenizar(oracion):
    tokens = word_tokenize(oracion.lower(), language="spanish")
    tokens_limpios = [t for t in tokens if t not in string.punctuation]
    return tokens_limpios

corpus = [limpiar_y_tokenizar(oracion) for oracion in oraciones]

# Entrenar modelo Word2Vec
modelo = Word2Vec(sentences=corpus, vector_size=50, window=2, min_count=1, workers=1)

# Palabras más similares a 'inteligencia'
print("Palabras similares a 'inteligencia':")
for palabra, similitud in modelo.wv.most_similar("inteligencia", topn=5):
    print(f"{palabra}: {similitud:.4f}")
````

# Word2Vec

## ¿Qué es Word2Vec?

**Word2Vec** es un modelo de red neuronal creado por Tomas Mikolov en Google (2013) para aprender representaciones vectoriales de palabras (embeddings). Estas representaciones capturan el significado, contexto y relaciones semánticas entre palabras.



## ¿Por qué es importante?

Modelos tradicionales como Bag of Words o TF-IDF:
- No capturan el significado ni el contexto.
- Generan vectores dispersos (sparse).
- No pueden entender similitud entre palabras.

**Word2Vec**:
- Representa palabras como vectores densos.
- Palabras similares → vectores similares.
- Captura semántica y sintaxis.



## ¿Cómo funciona Word2Vec?

### 1. CBOW (Continuous Bag of Words)
- Predice una palabra a partir de su contexto.
- Ejemplo:
  - Frase: "El gato duerme en la cama."
  - Contexto: [El, duerme, en, la]
  - Palabra objetivo: "gato"

### 2. Skip-gram
- Predice el contexto a partir de una palabra.
- Ejemplo:
  - Palabra central: "gato"
  - Contexto: [El, duerme, en, la]

Ambos usan una red neuronal con una sola capa oculta. El vector de esa capa es el **embedding**.



## ¿Qué aprende el modelo?

El modelo aprende relaciones semánticas como:
```
vec("rey") - vec("hombre") + vec("mujer") ≈ vec("reina")
```

También relaciones sintácticas:
```
vec("caminar") ≈ vec("caminando")
```



## Visualización

Word2Vec puede reducirse a 2D y mostrar agrupaciones semánticas de palabras similares. Por ejemplo:

- ["rey", "reina", "hombre", "mujer"] se agrupan según su rol y género.



## Requisitos

1. Corpus grande (Wikipedia, libros, noticias).
2. Preprocesamiento: limpieza, tokenización, etc.
3. Entrenamiento con `gensim` o frameworks.

O usar modelos preentrenados como:
- FastText (Facebook)
- SBWC (Spanish Billion Words Corpus)



## Limitaciones

- No maneja polisemia (palabras con múltiples significados).
- No toma en cuenta el contexto completo.
- Genera un solo vector por palabra (no contextual).

Modelos como **BERT** y **GPT** resuelven esto con representaciones contextuales.



## Comentarios

Word2Vec revolucionó el PLN al permitir que las máquinas entiendan las relaciones entre palabras de forma matemática. Aunque ha sido superado por modelos más avanzados, sigue siendo una base clave en el procesamiento de lenguaje natural.

````python
#Paso 1: Dataset y funciones auxiliares
import numpy as np

# Corpus de ejemplo
corpus =  """
La inteligencia artificial está revolucionando la forma en que interactuamos con la tecnología.
Muchas empresas utilizan algoritmos para mejorar sus servicios y productos.
El aprendizaje automático es una rama fundamental de la inteligencia artificial.
"""

# Tokenización y vocabulario
tokens = corpus.lower().split()
vocab = list(set(tokens))
vocab_size = len(vocab)
word_to_idx = {w: i for i, w in enumerate(vocab)}
idx_to_word = {i: w for w, i in word_to_idx.items()}

def one_hot_vector(word):
    vec = np.zeros(vocab_size)
    vec[word_to_idx[word]] = 1
    return vec
````

````python
#Paso 2: Generación de pares Skip-gram
def generate_skipgram_pairs(tokens, window_size=1):
    pairs = []
    for idx, word in enumerate(tokens):
        for offset in range(-window_size, window_size + 1):
            context_idx = idx + offset
            if context_idx >= 0 and context_idx < len(tokens) and context_idx != idx:
                pairs.append((word, tokens[context_idx]))
    return pairs

training_data = generate_skipgram_pairs(tokens, window_size=1)
````

````python
#Paso 3: Inicialización de pesos
embedding_dim = 5  # Tamaño del vector de embedding

# Matrices de pesos
W1 = np.random.rand(vocab_size, embedding_dim)
W2 = np.random.rand(embedding_dim, vocab_size)
````

````python
#Paso 4: Función de entrenamiento
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

def train(training_data, epochs=1000, lr=0.01):
    global W1, W2
    for epoch in range(epochs):
        loss = 0
        for target, context in training_data:
            x = one_hot_vector(target)  # Input
            y_true = one_hot_vector(context)  # Output esperada

            # Forward
            h = np.dot(W1.T, x)
            u = np.dot(W2.T, h)
            y_pred = softmax(u)

            # Cálculo de error
            e = y_pred - y_true

            # Backpropagation
            dW2 = np.outer(h, e)
            dW1 = np.outer(x, np.dot(W2, e))

            # Actualización
            W1 -= lr * dW1
            W2 -= lr * dW2

            loss += -np.log(y_pred[np.argmax(y_true)])
        if epoch % 100 == 0:
            print(f"Epoch {epoch}, loss: {loss:.4f}")
````

````python
#Paso 5: Entrenar
train(training_data, epochs=1000)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Corpus de ejemplo y vocabulario
corpus =  """
La inteligencia artificial está revolucionando la forma en que interactuamos con la tecnología.
Muchas empresas utilizan algoritmos para mejorar sus servicios y productos.
El aprendizaje automático es una rama fundamental de la inteligencia artificial.
"""
tokens = corpus.lower().split()
vocab = list(set(tokens))
vocab_size = len(vocab)
word_to_idx = {w: i for i, w in enumerate(vocab)}
idx_to_word = {i: w for w, i in word_to_idx.items()}

# One-hot encoding
def one_hot_vector(word):
    vec = np.zeros(vocab_size)
    vec[word_to_idx[word]] = 1
    return vec

# Generar pares skip-gram
def generate_skipgram_pairs(tokens, window_size=1):
    pairs = []
    for idx, word in enumerate(tokens):
        for offset in range(-window_size, window_size + 1):
            context_idx = idx + offset
            if context_idx >= 0 and context_idx < len(tokens) and context_idx != idx:
                pairs.append((word, tokens[context_idx]))
    return pairs

training_data = generate_skipgram_pairs(tokens, window_size=1)

# Inicializar pesos
embedding_dim = 5
W1 = np.random.rand(vocab_size, embedding_dim)
W2 = np.random.rand(embedding_dim, vocab_size)

# Funciones auxiliares
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

def train(training_data, epochs=1000, lr=0.01):
    global W1, W2
    for epoch in range(epochs):
        loss = 0
        for target, context in training_data:
            x = one_hot_vector(target)
            y_true = one_hot_vector(context)

            # Forward
            h = np.dot(W1.T, x)
            u = np.dot(W2.T, h)
            y_pred = softmax(u)

            # Error y retropropagación
            e = y_pred - y_true
            dW2 = np.outer(h, e)
            dW1 = np.outer(x, np.dot(W2, e))

            # Actualización
            W1 -= lr * dW1
            W2 -= lr * dW2

            loss += -np.log(y_pred[np.argmax(y_true)])
        if epoch % 200 == 0:
            print(f"Epoch {epoch}, Loss: {loss:.4f}")

# Entrenar
train(training_data, epochs=1000)

# Reducción de dimensionalidad con PCA
pca = PCA(n_components=2)
W1_reduced = pca.fit_transform(W1)

# Visualizar en 2D
plt.figure(figsize=(8, 6))
for i, word in enumerate(vocab):
    plt.scatter(W1_reduced[i, 0], W1_reduced[i, 1])
    plt.text(W1_reduced[i, 0] + 0.01, W1_reduced[i, 1] + 0.01, word)

plt.title("Visualización 2D de Word Embeddings entrenados")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
````

````python
import re

def encontrar_diptongos(texto):
    # Expresión regular para detectar diptongos comunes (simplificada)
    patron_diptongo = r'(?:[iu][aeo]|[aeo][iu]|i[u]|u[i])'

    # Buscar todos los diptongos en el texto
    diptongos = re.findall(patron_diptongo, texto.lower())

    return diptongos

# Ejemplo de uso
texto_ejemplo = "Cielo, tierra, viuda, aereo, cuidado, auto, ciudad."
resultado = encontrar_diptongos(texto_ejemplo)

print("Diptongos encontrados:", resultado)
````

````python
# prompt: Con expresión regular encontra un identificador valido para el nombre de una variable

# Expresión regular para encontrar un identificador válido para el nombre de una variable
# Un identificador válido en Python debe empezar con una letra (a-z, A-Z) o un guión bajo (_)
# seguido de cero o más letras, números (0-9) o guiones bajos.
# No puede empezar con un número.

# Definimos el patrón regex para un identificador válido
# ^       : Inicio de la cadena (opcional, para asegurar que coincide desde el principio si se testea una sola palabra)
# [a-zA-Z_] : El primer caracter debe ser una letra (mayúscula o minúscula) o un guión bajo
# [a-zA-Z0-9_]* : Los caracteres subsiguientes pueden ser letras, números o guiones bajos (cero o más veces)
# $       : Fin de la cadena (opcional, para asegurar que coincide hasta el final si se testea una sola palabra)
patron_identificador = r'^[a-zA-Z_][a-zA-Z0-9_]*$'

# Ejemplos de prueba
identificador1 = 'Aux'      # Válido
identificador2 = '3Aux'     # No válido (empieza con número)
identificador3 = '_mi_variable123' # Válido
identificador4 = 'otra-variable' # No válido (contiene guión medio)
identificador5 = 'for'      # Válido sintácticamente, pero es una palabra reservada (regex no valida si es reservada)
identificador6 = 'V!v_numero' # No válido (contiene !)


print(f"'{identificador1}': {'Válido' if re.match(patron_identificador, identificador1) else 'No válido'}")
print(f"'{identificador2}': {'Válido' if re.match(patron_identificador, identificador2) else 'No válido'}")
print(f"'{identificador3}': {'Válido' if re.match(patron_identificador, identificador3) else 'No válido'}")
print(f"'{identificador4}': {'Válido' if re.match(patron_identificador, identificador4) else 'No válido'}")
print(f"'{identificador5}': {'Válido' if re.match(patron_identificador, identificador5) else 'No válido'}")
print(f"'{identificador6}': {'Válido' if re.match(patron_identificador, identificador6) else 'No válido'}")

# También puedes usar search si quieres encontrar identificadores dentro de un texto más largo
texto_con_variables = "La variable Aux tiene un valor de 10. Pero 3Aux no es un nombre válido."
# Aquí usamos findall para encontrar todas las coincidencias del patrón en el texto
# La regex para findall podría ser ligeramente diferente si quieres extraerlos de un texto,
# pero para validar si una cadena *es* un identificador, re.match es más apropiado.
# Si quisieras extraer todos los identificadores de un texto, podrías usar findall con un patrón como r'[a-zA-Z_]\w*'
# \w es equivalente a [a-zA-Z0-9_]
patron_extracción = r'[a-zA-Z_]\w*'
identificadores_encontrados = re.findall(patron_extracción, texto_con_variables)
print(f"\nIdentificadores encontrados en el texto: {identificadores_encontrados}")
````

````python
#Crea una función que va a convertir el quijote en un dataset barajeado
def to_dataset(sequence, length, shuffle = False, seed = None, batch_size = 32):
    ds = tf.data.Dataset.from_tensor_slices(sequence)
    ds = ds.window(length + 1, shift = 1, drop_remainder = True)
    ds = ds.flat_map(lambda window_ds: window_ds.batch(length+1))
    if shuffle:
        ds = ds.shuffle(100_000, seed = seed)
    ds = ds.batch(batch_size)
    return ds.map(lambda window: (window[:,:-1],window[:,1:])).prefetch(1)
````

````python
#Las diferentes funciones de activación para redes neuronales
import numpy as np
import matplotlib.pyplot as plt

# 1. Función de Paso (Step Function / Heaviside)
def step_function(x):
    """
    Función de paso (Heaviside).
    Devuelve 1 si x > 0, de lo contrario devuelve 0.
    """
    return np.where(x > 0, 1, 0)

# 2. Función Sigmoide (Logística)
def sigmoid(x):
    """
    Función de activación Sigmoide (Logística).
    Comprime un valor real en el rango [0, 1].
    """
    return 1 / (1 + np.exp(-x))

# 3. Función ReLU (Rectified Linear Unit)
def relu(x):
    """
    Función de activación ReLU (Unidad Lineal Rectificada).
    Devuelve x si x > 0, de lo contrario devuelve 0.
    """
    return np.maximum(0, x)

# 4. Función Softmax
def softmax(x):
    """
    Función de activación Softmax.
    Transforma un vector de valores reales en una distribución de probabilidad.
    """
    e_x = np.exp(x - np.max(x))  # Resta np.max(x) para estabilidad numérica
    return e_x / e_x.sum(axis=0)

# Generar un rango de valores para graficar
x_values = np.linspace(-5, 5, 100) # Para Sigmoide, Step y ReLU
x_softmax = np.array([1.0, 2.0, 3.0]) # Ejemplo de vector para Softmax

# Calcular las salidas de las funciones de activación
y_step = step_function(x_values)
y_sigmoid = sigmoid(x_values)
y_relu = relu(x_values)
y_softmax = softmax(x_softmax)

# Visualizar las funciones de activación
plt.figure(figsize=(15, 10))

# Gráfico de la Función de Paso
plt.subplot(2, 2, 1)
plt.plot(x_values, y_step, label='Función de Paso')
plt.title('Función de Paso (Heaviside)')
plt.xlabel('x')
plt.ylabel('Salida')
plt.grid(True)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(1, color='gray', linestyle='--', linewidth=0.8)
plt.yticks([0, 1])

# Gráfico de la Función Sigmoide
plt.subplot(2, 2, 2)
plt.plot(x_values, y_sigmoid, label='Función Sigmoide', color='orange')
plt.title('Función Sigmoide (Logística)')
plt.xlabel('x')
plt.ylabel('Salida')
plt.grid(True)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(0.5, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(1, color='gray', linestyle='--', linewidth=0.8)

# Gráfico de la Función ReLU
plt.subplot(2, 2, 3)
plt.plot(x_values, y_relu, label='Función ReLU', color='green')
plt.title('Función ReLU (Rectified Linear Unit)')
plt.xlabel('x')
plt.ylabel('Salida')
plt.grid(True)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)

# Mostrar ejemplo de Softmax (no es una función de una sola entrada vs. salida)
plt.subplot(2, 2, 4)
bars = plt.bar(['x1', 'x2', 'x3'], y_softmax, color=['red', 'purple', 'blue'])
plt.title('Ejemplo de Función Softmax')
plt.xlabel('Entradas')
plt.ylabel('Probabilidad de Salida')
plt.grid(axis='y')
plt.ylim(0, 1)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.02, round(yval, 3), ha='center', va='bottom')
plt.text(1.5, 0.8, f'Entradas originales: {x_softmax}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
plt.text(1.5, 0.75, f'Suma de probabilidades: {np.sum(y_softmax):.3f}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)


plt.tight_layout()
plt.show()

print("\n--- Demostración de valores para cada función ---")
print(f"Valores de entrada (x): {np.round(x_values[[0, 25, 50, 75, 99]], 2)}")
print(f"Función de Paso para x: {step_function(np.array([-2, -0.1, 0, 0.1, 2]))}")
print(f"Sigmoide para x: {np.round(sigmoid(np.array([-2, -0.1, 0, 0.1, 2])), 3)}")
print(f"ReLU para x: {relu(np.array([-2, -0.1, 0, 0.1, 2]))}")
print(f"\nEntradas para Softmax: {x_softmax}")
print(f"Salidas Softmax: {np.round(y_softmax, 3)}")
print(f"Suma de salidas Softmax: {np.sum(y_softmax):.3f}")
````

````python
#Creamos una función llamada next char para crear texto nuevo
def next_char(text, temperature = 1):
    y_proba = quijote_modelo.predict(tf.constant([text]))[0,-1:]
    rescaled_logits = tf.math.log(y_proba)/temperature
    char_id = tf.random.categorical(rescaled_logits, num_samples = 1)[0,0]
    return text_vec_layer.get_vocabulary()[char_id+2]
````

````python
#Creamos una función llamada extend_text para crear la secuencia completa
def extend_text(text, n_chars = 50, temperature = 1):
    for _ in range(n_chars):
        text += next_char(text, temperature)
    return text
````

````python
#Preparar una capa de embedding cada uno(decoder y encoder)
encoder_embedding_layer = tf.keras.layers.Embedding(vocab_size, embed_size, mask_zero = True)
decoder_embedding_layer = tf.keras.layers.Embedding(vocab_size, embed_size, mask_zero=True)
````

````python
#Ahora coloca tus input_ids a la capa de embedding
encoder_embeddings = encoder_embedding_layer(encoder_input_ids)
decoder_embeddings = decoder_embedding_layer(decoder_input_ids)
````

````python
#Definir el encoder como 1 sola capa LSTM de 512 neuronas, pasa los embeddings por esa capa salen los outputs y el state
encoder = tf.keras.layers.LSTM(512, return_state = True)
encoder_outputs, *encoder_state = encoder(encoder_embeddings)
````

````python
#Repitir el proceso correspondiente para el decoder
decoder = tf.keras.layers.LSTM(512, return_sequences = True)
decoder_outputs = decoder(decoder_embeddings, initial_state = encoder_state)
````

````python
#La capa de salida será una capa Densa del mismo tamaño que el vocabulario con activacion softmax, da la probabilidad Y de alguna palabra tomando los decoder outputs como entrada
output_layer = tf.keras.layers.Dense(vocab_size, activation="softmax")
Y_proba = output_layer(decoder_outputs)
````

````python
# Armar el modelo
model = tf.keras.Model(inputs = [encoder_inputs, decoder_inputs], outputs = [Y_proba])
````

````python
#Compila
model.compile(loss="sparse_categorical_crossentropy", optimizer="nadam",
              metrics=["accuracy"])
````

````python
#Arma un checkpoint para guardar el modelo
tmodel_ckpt = tf.keras.callbacks.ModelCheckpoint(
    "traductor_modelo.keras", monitor="val_accuracy", save_best_only=True)
````

````python
#Ejecuta a 10 epocas, nota que tienes que meter 2 sets de entrenamiento y validación para X (enc y dec) - se va a tardar como 20 minutos por epoca sin GPU!
history=model.fit((X_train, X_train_dec), Y_train, epochs = 10, validation_data=((X_valid, X_valid_dec), Y_valid), callbacks = [tmodel_ckpt])
````

````python
# Ahora a armar el traductor
def translate(sentence_en):
    translation = ""
    for word_idx in range(max_length):
        # Pass the string tensors directly to the model
        y_proba = model.predict((tf.constant([sentence_en]), tf.constant(["startofseq" + translation])))[0,word_idx]
        predicted_word_id=np.argmax(y_proba)
        predicted_word =text_vec_layer_es.get_vocabulary()[predicted_word_id]
        if predicted_word == "endofseq":
            break
        translation += " " + predicted_word
    return translation.strip()
````

````python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# 1. Dataset de ejemplo (texto y sentimiento)
data = [
    ("Me encantó la película, fue maravillosa", "positivo"),
    ("Una experiencia realmente mala", "negativo"),
    ("Excelente atención y muy buena comida", "positivo"),
    ("No me gustó el servicio", "negativo"),
    ("El producto es fantástico, lo recomiendo", "positivo"),
    ("Demasiado lento y aburrido", "negativo")
]

# 2. Separar datos y etiquetas
textos, etiquetas = zip(*data)

# 3. Vectorizar el texto (Bag of Words)
vectorizador = CountVectorizer()
X = vectorizador.fit_transform(textos)
y = etiquetas

# 4. Separar entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# 5. Entrenar modelo Naive Bayes
modelo = MultinomialNB()
modelo.fit(X_train, y_train)

# 6. Evaluar modelo
y_pred = modelo.predict(X_test)
print("Reporte de clasificación:\n")
print(classification_report(y_test, y_pred))
````

````python
import torch
from torch import nn
import torch.nn.functional as F
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
from torch.utils.data import DataLoader
from torch.nn.utils.rnn import pad_sequence

````

````python
# Datos de ejemplo (paralelos)
data = [
    ("hola", "hello"),
    ("buenos días", "good morning"),
    ("¿cómo estás?", "how are you?"),
    ("gracias", "thank you"),
    ("hasta luego", "see you later"),
]

# Tokenizadores
SRC_LANG = 'es'
TGT_LANG = 'en'
tokenizer_es = get_tokenizer('spacy', language='es_core_news_sm')
tokenizer_en = get_tokenizer('spacy', language='en_core_web_sm')

# Vocabularios
def yield_tokens(data_iter, tokenizer):
    for src, tgt in data_iter:
        yield tokenizer(src)
        yield tokenizer(tgt)

vocab_src = build_vocab_from_iterator((tokenizer_es(src) for src, _ in data), specials=['<unk>', '<pad>', '<bos>', '<eos>'])
vocab_src.set_default_index(vocab_src['<unk>'])
vocab_tgt = build_vocab_from_iterator((tokenizer_en(tgt) for _, tgt in data), specials=['<unk>', '<pad>', '<bos>', '<eos>'])
vocab_tgt.set_default_index(vocab_tgt['<unk>'])

# Funciones de preprocesamiento
def tensor_transform(token_ids): return torch.cat((torch.tensor([2]), torch.tensor(token_ids), torch.tensor([3])))

def collate_fn(batch):
    src_batch, tgt_batch = [], []
    for src_sample, tgt_sample in batch:
        src_tokens = vocab_src(tokenizer_es(src_sample))
        tgt_tokens = vocab_tgt(tokenizer_en(tgt_sample))
        src_batch.append(tensor_transform(src_tokens))
        tgt_batch.append(tensor_transform(tgt_tokens))
    src_batch = pad_sequence(src_batch, padding_value=1)
    tgt_batch = pad_sequence(tgt_batch, padding_value=1)
    return src_batch, tgt_batch

dataloader = DataLoader(data, batch_size=50, collate_fn=collate_fn)

# Modelo Transformer
class Seq2SeqTransformer(nn.Module):
    def __init__(self, num_encoder_layers, num_decoder_layers, emb_size, nhead, src_vocab_size, tgt_vocab_size, dim_feedforward=512):
        super().__init__()
        self.transformer = nn.Transformer(d_model=emb_size, nhead=nhead,
                                          num_encoder_layers=num_encoder_layers,
                                          num_decoder_layers=num_decoder_layers,
                                          dim_feedforward=dim_feedforward)
        self.src_tok_emb = nn.Embedding(src_vocab_size, emb_size)
        self.tgt_tok_emb = nn.Embedding(tgt_vocab_size, emb_size)
        self.positional_encoding = nn.Parameter(torch.rand(5000, emb_size), requires_grad=True)
        self.generator = nn.Linear(emb_size, tgt_vocab_size)

    def forward(self, src, tgt):
        src_emb = self.src_tok_emb(src) + self.positional_encoding[:src.size(0), :]
        tgt_emb = self.tgt_tok_emb(tgt) + self.positional_encoding[:tgt.size(0), :]
        output = self.transformer(src_emb, tgt_emb)
        return self.generator(output)

# Instanciar el modelo
SRC_VOCAB_SIZE = len(vocab_src)
TGT_VOCAB_SIZE = len(vocab_tgt)
EMB_SIZE = 128
NHEAD = 4
FFN_HID_DIM = 256
NUM_ENCODER_LAYERS = 3
NUM_DECODER_LAYERS = 3

model = Seq2SeqTransformer(NUM_ENCODER_LAYERS, NUM_DECODER_LAYERS, EMB_SIZE, NHEAD, SRC_VOCAB_SIZE, TGT_VOCAB_SIZE, FFN_HID_DIM)

# Ejemplo de uso
src_text = "Hoy es miercoles en la noche"
src_tokens = tensor_transform(vocab_src(tokenizer_es(src_text))).unsqueeze(1)  # (seq_len, batch_size)
tgt_tokens = tensor_transform(vocab_tgt(tokenizer_en("thank you"))[:-1]).unsqueeze(1)

with torch.no_grad():
    out = model(src_tokens, tgt_tokens)
    out_tokens = out.argmax(-1).squeeze()

print("Entrada:", src_text)
flat_out_tokens = out_tokens.view(-1).tolist()  # Asegura que sea una lista plana
print("Traducción aproximada:", ' '.join(vocab_tgt.lookup_tokens(flat_out_tokens)))
````
