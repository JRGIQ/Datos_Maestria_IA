```python
# 2. Configura tu clave de API
from getpass import getpass
import os

openai_api_key = getpass("Pega aquí tu OPENAI_API_KEY: ")
os.environ["OPENAI_API_KEY"] = openai_api_key
```

```python
# 3. Usa el nuevo cliente OpenAI
from openai import OpenAI
import os

# Inicializa el cliente con tu clave
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def chat_with_gpt(prompt: str,
                  model: str = "gpt-3.5-turbo",
                  temperature: float = 0.5,
                  max_tokens: int = 500) -> str:
    """
    Envía un prompt a la API de ChatGPT (nueva versión) y devuelve la respuesta.
    """
    resp = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "Eres un asistente experto en Inteligencia Artificial."},
            {"role": "user",   "content": prompt},
        ],
        temperature=temperature,
        max_tokens=max_tokens,
    )
    # La respuesta viene en resp.choices[0].message.content
    return resp.choices[0].message.content.strip()
```

```python
# 4. Prueba tu función
if __name__ == "__main__":
    pregunta = "Explica el concepto de A*"
    respuesta = chat_with_gpt(pregunta)
    print("ChatGPT responde:\n", respuesta)
```
